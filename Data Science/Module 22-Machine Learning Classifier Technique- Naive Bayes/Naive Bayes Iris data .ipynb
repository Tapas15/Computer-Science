# Train test splitting before model making 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Species',axis=1)# dropping the catagorical dummy variable
y = df['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

### Naive bayes has 3 models lets impport the models and testify each area 

# Gaussian classifier naive bayes 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
ignb = GaussianNB()
```

```python
ignb.fit(X_train,y_train) # Gaussian classifier fitting model in 
```

# Prediction and evaluation 

```python
predictions = ignb.predict(X_test)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

# Multinomial Classification 

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
imnb = MultinomialNB()
```

```python
imnb.fit(X_train,y_train) # multinomial model
```

# Prediction and Evaluation 

```python
predictions = imnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

# Bernauli Classifier naive bayes

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
ibnb = BernoulliNB()
```

```python
ibnb.fit(X_train,y_train) # model making 
```

# Prediction and evaluation 

```python
predictions = ibnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

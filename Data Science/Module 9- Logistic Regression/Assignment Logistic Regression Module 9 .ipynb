### A Brief model before train test split 

```r
model  <- glm((naffairs >0)~kids+unhap+vryunhap+unhap+avgmarr+hapavg+vryhap+antirel+notrel+slghtrel,data = df,family = binomial)
```

```r
plot(model)
```

```r
exp(coef(model))
```

### lets make our model train test split of data 

```r
library(caTools)# will use for split our data 
set.seed(101)

split = sample.split(df$naffairs, SplitRatio = 0.70)

final.train = subset(df, split == TRUE)
final.test = subset(df, split == FALSE)
```

```r
final_glm_modle  <- glm((naffairs>0)~.,family = binomial,data=final.train) # final model included all variable 
```

```r
plot(final_glm_modle)
```

```r
print(summary(final_glm_modle)) #  in our model we can see the values which is not fitting
```

### Lets Check our probablity and prediction accuracy

```r
final_glm_modle1  <- glm((naffairs>0)~.-vryrel-vryhap-yrsmarr6,family = binomial,data=final.train) # final model included all variable 
```

```r
print(summary(final_glm_modle1))
```

```r
fitted.probabilities <- predict(final_glm_modle1,newdata=final.test,type='response')
```

```r
head(fitted.probabilities)
```

```r
fitted.results <- ifelse(fitted.probabilities > 0.5,1,0)
yes_no  <- ifelse(fitted.probabilities > 0.5, "Yes","No")
```

```r
head(fitted.results) # will give result 0 or 1
```

```r
misClasificError <- mean(fitted.results != final.test$naffairs)
```

### Accuracy sencitivity and pecificity¶

```r
print(paste('Accuracy',1-misClasificError)) # accuracy of our model is not so good
```

```r
threshold=0.5
predicted_values<-ifelse(predict(final_glm_modle1,type="response")>threshold,1,0)
actual_values<-final_glm_modle1$y
conf_matrix<-table(predicted_values,actual_values)
conf_matrix
```

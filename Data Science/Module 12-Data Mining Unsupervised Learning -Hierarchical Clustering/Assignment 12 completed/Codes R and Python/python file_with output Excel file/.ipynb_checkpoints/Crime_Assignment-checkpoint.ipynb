```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())	/	(i.max()	-	i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,-4:])
df_norm.head() # running noramalization function of data set
```

```python
df_norm.head(5) # normal head is coming
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
```

```python
type(df_norm)
z = linkage(df_norm, method="complete",metric="euclidean")
```

# Model building 

```python
plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
z1 = linkage(df_norm, method="ward",metric="euclidean")
```

```python
plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z1,
    leaf_rotation=0.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
# Now applying AgglomerativeClustering choosing 3 as clusters from the dendrogram
from	sklearn.cluster	import	AgglomerativeClustering 
h_complete	=	AgglomerativeClustering(n_clusters=2,	linkage='ward',affinity = "euclidean").fit(df_norm) 
```

```python
cluster_labels=pd.Series(h_complete.labels_)
crime['clust']=cluster_labels # creating a  new column and assigning it to new column 
#univ = univ.iloc[:,[8,7,0,1,2,3,4,5,6]]
crime.head()
```

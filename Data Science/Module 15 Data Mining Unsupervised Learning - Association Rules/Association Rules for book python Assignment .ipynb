# Association Rules for book python Assignment 

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
#import matplotlib.pyplot as plt
```

```python
book = pd.read_csv('book.csv')
```

```python
book # already rule formatted 
```

```python
frequent_itemsets = apriori(book, min_support=0.005, max_len=3,use_colnames = True)
```

```python
frequent_itemsets # supported frequent items 
```

```python
Arules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
############################## Model Buiding############################################ 
import numpy as np
#import statsmodels.api as sm
import statsmodels.formula.api as smf
from plotnine import *
(ggplot(wcat,aes("Waist","AT"))+geom_point())+stat_smooth(method = 'lm') # Scattarplot and with abline 
```

```python
##### MODEL BUILDING ########
#model 1 
plt.scatter(x=wcat['Waist'], y=wcat['AT'],color='green')# Scatter plot
np.corrcoef(wcat.Waist, wcat.AT) #correlation
mod = smf.ols('AT ~ Waist', data=wcat).fit()
mod.summary() # to see the summary of out model

pred1 = mod.predict(pd.DataFrame(wcat['Waist'])) # model prediction 
pred1
#printing level of interval for confidence
print (mod.conf_int(0.01)) # 99% confidence interval
# residuals
res = wcat.AT - pred1
res
#rmse
sqres = res*res
mse = np.mean(sqres)
rmse = np.sqrt(mse)
rmse
```

```python
######### Model building on Transformed Data

# Log Transformation
# x = log(waist); y = at
plt.scatter(x=np.log(wcat['Waist']),y=wcat['AT'],color='brown')
np.corrcoef(np.log(wcat.Waist), wcat.AT) #correlation
# linear model 2 
model2 = smf.ols('AT ~ np.log(Waist)',data=wcat).fit()
model2.summary()
# prediction
pred2 = model2.predict(pd.DataFrame(wcat['Waist']))
pred2
# printing inter val level of confidence 
print(model2.conf_int(0.01)) # 99% confidence level
# residuals
res2 = wcat.AT - pred2
res2
# rmse of second model
sqres2 = res2*res2
mse2 = np.mean(sqres2)
rmse2 = np.sqrt(mse2)
rmse2
```

```python
pred2 = model2.predict(pd.DataFrame(wcat['Waist']))
pred2
```

```python
# Exponential transformation
plt.scatter(x=wcat['Waist'], y=np.log(wcat['AT']),color='orange')

np.corrcoef(wcat.Waist, np.log(wcat.AT)) #correlation

model3 = smf.ols('np.log(AT) ~ Waist',data=wcat).fit()
model3.summary()
# predictions
pred_log = model3.predict(pd.DataFrame(wcat['Waist']))
pred_log
pred3 = np.exp(pred_log)
pred3
# printing the confidence interval
print(model3.conf_int(0.01)) # 99% confidence level
# residuals 
res3 = wcat.AT - pred3
res3
#rmse of third 
sqres3 = res3*res3
mse3 = np.mean(sqres3)
rmse3 = np.sqrt(mse3)
rmse3

```

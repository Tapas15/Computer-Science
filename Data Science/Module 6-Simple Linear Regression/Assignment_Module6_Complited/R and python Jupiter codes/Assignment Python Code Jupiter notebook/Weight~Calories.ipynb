```python
############################## Model Buiding############################################ 
import numpy as np
#import statsmodels.api as sm
import statsmodels.formula.api as smf
from plotnine import *
(ggplot(cal,aes("Calories","Weight"))+geom_point())+stat_smooth(method = 'lm') # Scattarplot and with abline 
```

##### MODEL BUILDING ########

```python
#model 1 
plt.scatter(x=cal['Calories'], y=cal['Weight'],color='green')# Scatter plot
np.corrcoef(cal.Calories, cal.Weight) #correlation
mod = smf.ols('Weight ~ Calories', data=cal).fit()
mod.summary() # to see the summary of out model

```

```python
pred1 = mod.predict(pd.DataFrame(cal['Calories'])) # model prediction 
pred1
```

```python
#printing level of interval for confidence
print (mod.conf_int(0.01)) # 99% confidence interval
```

```python
# residuals
res = cal.Weight - pred1
res
```

```python
#rmse
sqres = res*res
mse = np.mean(sqres)
rmse = np.sqrt(mse)
rmse
```

```python
# model 2 

######### Model building on Transformed Data

# Log Transformation
# x = log(waist); y = at
plt.scatter(x=np.log(cal['Calories']),y=cal['Weight'],color='brown')
np.corrcoef(np.log(cal.Calories), cal.Weight) #correlation
# linear model 2 
model2 = smf.ols('Weight ~ np.log(Calories)',data=cal).fit()
model2.summary()
```

```python
# prediction
pred2 = model2.predict(pd.DataFrame(cal['Calories']))
pred2
```

```python
# printing inter val level of confidence 
print(model2.conf_int(0.01)) # 99% confidence level
```

```python
# residuals
res2 = cal.Weight - pred2
res2
```

```python
# rmse of second model
sqres2 = res2*res2
mse2 = np.mean(sqres2)
rmse2 = np.sqrt(mse2)
rmse2
```

```python
# Exponential transformation
plt.scatter(x=cal['Calories'], y=np.log(cal['Weight']),color='orange')

np.corrcoef(cal.Calories, np.log(cal.Weight)) #correlation

model3 = smf.ols('np.log(Weight) ~ Calories',data=cal).fit()
model3.summary()
```

```python
# predictions
pred_log = model3.predict(pd.DataFrame(cal['Calories']))
pred_log
pred3 = np.exp(pred_log)
pred3
```

```python
# printing the confidence interval
print(model3.conf_int(0.01)) # 99% confidence level
```

```python
# residuals 
res3 = cal.Weight - pred3
res3
```

```python
#rmse of third 
sqres3 = res3*res3
mse3 = np.mean(sqres3)
rmse3 = np.sqrt(mse3)
rmse3
```
